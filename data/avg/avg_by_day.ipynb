```python
import pandas as pd
```

```python
data = pd.read_csv('fixed_data.csv')
data.head()
```

```python
df_mazanovo = data[['Unnamed: 0', 'Value-316-zeya-mazanovo']]
df_malayasazanka = data[['Unnamed: 0', 'Value-1002-zeya-malayasazanka']]
df_svobodny = data[['Unnamed: 0', 'Value-1003-zeya-svobodny']]
df_blagoveshensk = data[['Unnamed: 0', 'Value-2366-zeya-blagoveshensk']]
df_belogorye = data[['Unnamed: 0', 'Value-317-zeya-belogorye']]
df_belogorye
```

```python
def average(glist):
    return sum(glist)/len(glist)
```

```python
from pprint import pprint
#объединить в список значения по дням
np_mazanovo = df_mazanovo.to_numpy()
np_mazanovo[0][0].split('-')

val = []
avg_by_day = {}
for el in np_mazanovo:
#     print(el)
    splited_data = el[0].split('-')
    day = splited_data[2]
    month = splited_data[1]
    value = el[1]
    data_key = month + '-' + day
#     print(f"{day} {month}")
#     if day == '01' and month == '01':
#         val.append(el[1])
        
    if data_key in avg_by_day:
        avg_by_day[data_key].append(value)
    else:
        avg_by_day[data_key] = [value]
    

for el in avg_by_day:
#     print(avg_by_day[el])
    avg_by_day[el] = average(avg_by_day[el])
#     print(avg_by_day[el])
#     break
# pprint(avg_by_day)

df_avg_by_day_mazanovo = pd.DataFrame(avg_by_day.items(), columns=['date', 'value-mazanovo'])
df_avg_by_day_mazanovo
```
