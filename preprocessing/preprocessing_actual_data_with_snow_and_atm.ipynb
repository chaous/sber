```python
import pandas as pd

df_atm = pd.read_csv('atmosf_rains.csv')
df_atm = df_atm.drop(['0.2', '0.4', '0.5', '2008', '1', '1.1', '0', '31510'], axis=1)
df_atm = df_atm[df_atm['2008.1'] < 2021]
df_atm = df_atm.drop_duplicates(subset=['2008.1', '1.2', '1.3'])
df_atm
```

```python
df = pd.read_csv('fixed_data_with_snow.csv')
df = df.drop(['Unnamed: 0'], axis=1)
for name in df.iloc[0]:
    print(name)
```

```python
rain_dic = {}
from tqdm import tqdm
for collum in df_atm:
    if collum != '2008.1' or collum != '1.2' or collum != '1.3':
        rain_dic[collum] = []
for date in tqdm(df['Timestamp']):
    year = int(date[0:4])
    month = int(date[5:7])
    day = int(date[8:10])
    working_dates = df_atm[df_atm['1.3'] >= day]
    working_dates = working_dates[working_dates['1.2'] >= month]
    working_dates = working_dates[working_dates['2008.1'] >= year]
    #working_dates = working_dates.iloc[:1 , :]
    if working_dates.empty:
        working_dates = working_dates[df_atm['1.2'] >= month]
        working_dates = working_dates[working_dates['2008.1'] >= year]
    if working_dates.empty:
        working_dates = working_dates[df_atm['2008.1'] >= year]
    if working_dates.empty:
        working_dates = df_atm.iloc[-2:-1 , :]
    working_dates = working_dates.iloc[:1 , :]
    for collum in df_atm:
        if collum != '2008.1' or collum != '1.2' or collum != '1.3':
            rain_dic[collum].append(working_dates.iloc[0][collum])
for i in rain_dic:
    df[i] = rain_dic[i]
df
```
