```python
import pandas as pd
```

```python
def Average(l): 
    avg = sum(l) / len(l) 
    return avg
```

```python
path = 'fixed_data.csv'
data = pd.read_csv(path)
```

```python
data.head()
```

```python
df_malayasazanka = data[['Unnamed: 0', 'Value-1002-zeya-malayasazanka']]
df_svobodny = data[['Unnamed: 0', 'Value-1003-zeya-svobodny']]
df_blagoveshensk = data[['Unnamed: 0', 'Value-2366-zeya-blagoveshensk']]
df_mazanovo = data[['Unnamed: 0', 'Value-316-zeya-mazanovo']]
df_belogorye = data[['Unnamed: 0', 'Value-317-zeya-belogorye']]
```

```python
df_malayasazanka_np = df_belogorye.to_numpy()

val_by_years = {}
for el in df_malayasazanka_np:
    year_tmp = el[0].split('-')[0]
    value_tmp = el[1]
    if year_tmp in val_by_years:
        val_by_years[year_tmp].append(value_tmp)
    else:
        val_by_years[year_tmp] = [value_tmp]
val_by_years
```

```python


avg_over_years = {}
for el in val_by_years:
    avg_over_years[el] = Average(val_by_years[el]) 
    

print('df_belogorye')
avg_over_years
```
